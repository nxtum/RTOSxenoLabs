```python
import os
import json
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

jsonFolder = '/content/json_output'
inputLen = 200
predLen = 100
stride = 100
numEpochs = 5
batchSize = 64
generateLen = 5000
temperature = 1.0
```

```python
db = {'train': [], 'valid': []}
for split in db:
    path = os.path.join(jsonFolder, split)
    if not os.path.exists(path): continue
    for file in os.listdir(path):
        if file.endswith(".json"):
            try:
                with open(os.path.join(path, file)) as f:
                    song = json.load(f)
                    if song and song[0][0] == "clock":
                        song = song[1:]
                    db[split].append(song)
            except:
                pass

def sliceSongs(songs, inLen, outLen, step):
    x, y = [], []
    for song in songs:
        if len(song) < inLen + outLen:
            continue
        for i in range(0, len(song) - inLen - outLen + 1, step):
            x.append(song[i:i + inLen])
            y.append(song[i + inLen:i + inLen + outLen])
    return x, y

def encode(x, y, cmdToId):
    xOut, yOut = [], []
    for a, b in zip(x, y):
        aIds = [cmdToId.get(tuple(i), -1) for i in a]
        bIds = [cmdToId.get(tuple(i), -1) for i in b]
        if -1 not in aIds and -1 not in bIds:
            xOut.append(aIds)
            yOut.append(bIds)
    return xOut, yOut
```

```python
trainX, trainY = sliceSongs(db['train'], inputLen, predLen, stride)
validX, validY = sliceSongs(db['valid'], inputLen, predLen, stride)
trainX, trainY = trainX[:20000], trainY[:20000]
validX, validY = validX[:2000], validY[:2000]

vocab = set(tuple(c) for song in (trainX + validX + trainY + validY) for c in song)
cmdToId = {cmd: i for i, cmd in enumerate(sorted(vocab))}
idToCmd = {i: cmd for cmd, i in cmdToId.items()}
vocabSize = len(cmdToId)

trainXEnc, trainYEnc = encode(trainX, trainY, cmdToId)
validXEnc, validYEnc = encode(validX, validY, cmdToId)
```

```python
class MusicDataset(Dataset):
    def __init__(self, x, y):
        self.x, self.y = x, y
    def __len__(self): return len(self.x)
    def __getitem__(self, idx):
        return torch.tensor(self.x[idx], dtype=torch.long), torch.tensor(self.y[idx], dtype=torch.long)

trainLoader = DataLoader(MusicDataset(trainXEnc, trainYEnc), batch_size=batchSize, shuffle=True)
validLoader = DataLoader(MusicDataset(validXEnc, validYEnc), batch_size=batchSize)

class BiLSTM(nn.Module):
    def __init__(self, vocabSize):
        super().__init__()
        self.embedding = nn.Embedding(vocabSize, 256)
        self.lstm = nn.LSTM(256, 512, 2, batch_first=True, bidirectional=True, dropout=0.3)
        self.norm = nn.LayerNorm(1024)
        self.drop = nn.Dropout(0.3)
        self.fc = nn.Linear(1024, vocabSize)
    def forward(self, x):
        x = self.embedding(x)
        x, _ = self.lstm(x)
        x = self.norm(x)
        x = self.drop(x)
        return self.fc(x[:, -predLen:, :])
```

```python
model = BiLSTM(vocabSize).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()


trainLosses, valLosses, trainAccs, valAccs = [], [], [], []
for epoch in range(numEpochs):
    model.train()
    totalLoss, correct, total = 0, 0, 0
    for x, y in trainLoader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        out = model(x)
        loss = criterion(out.view(-1, vocabSize), y.view(-1))
        loss.backward()
        optimizer.step()
        totalLoss += loss.item()
        correct += (out.argmax(2) == y).sum().item()
        total += y.numel()
    trainLosses.append(totalLoss / len(trainLoader))
    trainAccs.append(correct / total)

    model.eval()
    valLoss, correct, total = 0, 0, 0
    with torch.no_grad():
        for x, y in validLoader:
            x, y = x.to(device), y.to(device)
            out = model(x)
            valLoss += criterion(out.view(-1, vocabSize), y.view(-1)).item()
            correct += (out.argmax(2) == y).sum().item()
            total += y.numel()
    valLosses.append(valLoss / len(validLoader))
    valAccs.append(correct / total)

    print(f"Epoch {epoch+1}/{numEpochs}: Loss {trainLosses[-1]:.4f} | Val {valLosses[-1]:.4f} | Acc {trainAccs[-1]:.4f} | ValAcc {valAccs[-1]:.4f}")
```

```python
def sampleWithTemp(logits, temp=1.0):
    probs = F.softmax(logits / temp, dim=-1)
    return torch.multinomial(probs, 1)

def generateSequence(model, startSeq, totalLen, temp=1.0):
    model.eval()
    result = startSeq.copy()
    while len(result) < totalLen:
        x = torch.tensor([result[-inputLen:]], dtype=torch.long).to(device)
        with torch.no_grad():
            logits = model(x)[0]
            nextTokens = sampleWithTemp(logits, temp).squeeze().tolist()
            result.extend(nextTokens)
    return result[:totalLen]

def saveNeslm(commands, file, clock=['clock', 1789773]):
    with open(file, 'w') as f:
        f.write(f"{clock[0]},{clock[1]}\n")
        for token in commands:
            cmd = idToCmd[token]
            f.write(f"{cmd[0]},{cmd[1]}\n")

for i in range(3):
    start = trainXEnc[i][:inputLen]
    gen = generateSequence(model, start, generateLen, temperature)
    saveNeslm(gen, f"generated_song_{i+1}.neslm")

def plotMetric(train, val, title, label):
    plt.figure()
    plt.plot(train, label='Train')
    plt.plot(val, label='Valid')
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel(label)
    plt.legend()
    plt.grid(True)
    plt.show()

plotMetric(trainLosses, valLosses, "Cross Entropy Loss", "Loss")
plotMetric(trainAccs, valAccs, "Accuracy", "Accuracy")
```
