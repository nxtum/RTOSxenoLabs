```python
path_to_first = "/content/nesmdb_nlm/test/002_1943_TheBattleofMidway_03_04AirBattleA.nlm.pkl"
with open(path_to_first, 'rb') as f:
    commands = pickle.load(f)
```

```python
_, rate = commands[0]
commands = commands[1:]

functions = list(filter(lambda x: x[0] != 'w', commands)) # all nonwait instructions
waits = list(filter(lambda x: x[0] == 'w', commands)) # wait instructions

print('NES CPU rate: {}'.format(rate))
print('Number of instructions: {}'.format(len(functions)))
print('Length of VGM: {}'.format(sum([x[1] for x in waits]) / 44100.))

for cmd in commands[:10]:
    print(cmd)
```
