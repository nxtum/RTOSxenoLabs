```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# EXAMPLE OF LSTM
# https://github.com/wdomski/AISELLab

def generate_series(freq_list, offset_list, amp_list, length=100, n=1000, series=1, noise=0.2):
    t = np.linspace(0, length, n)
    data = np.zeros((n,))
    for i in range(len(freq_list)):
        freq = freq_list[i]
        offset = offset_list[i]
        amp = amp_list[i]
        data += amp * np.sin(2 * np.pi * (t - offset) * freq)
    data = data + noise * (np.random.rand(series, n) - 0.5)
    return t, data[..., np.newaxis].astype(np.float32)


def moving_window(data, window_size=25, y_size=1):
    n = data.shape[1] // (window_size + y_size)
    size = int(n * (window_size + y_size))
    reshaped = data[0, :size, 0].reshape(n, window_size + y_size, 1)
    x = reshaped[:, 0:window_size]
    y = reshaped[:, window_size:window_size + y_size]
    return x, y

input_size = 100
predicted_steps = 100
time, series = generate_series([1, 2, 10], [0, 0.3, 3], [1, 1, 0.2],
                               length=40000, n=1000000, series=1, noise=0.2)

x, y = moving_window(series, window_size=input_size, y_size=predicted_steps)

train_ind = int(len(x) * 0.8)
x_train, y_train = x[:train_ind], y[:train_ind]
x_test, y_test = x[train_ind:], y[train_ind:]
lstm_model = tf.keras.Sequential([
    tf.keras.Input(shape=(input_size, 1)),
    tf.keras.layers.LSTM(64, return_sequences=True),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1)
])
lstm_model.compile(loss='mse', optimizer='adam', metrics=['mae'])

lstm_h = lstm_model.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=10, batch_size=64)

plt.figure(figsize=(12, 6))
plt.plot(lstm_h.history['loss'], label='Training Loss', linestyle='--', color='blue')
plt.plot(lstm_h.history['val_loss'], label='Validation Loss', color='blue')
plt.plot(lstm_h.history['mae'], label='Training MAE', linestyle='--', color='green')
plt.plot(lstm_h.history['val_mae'], label='Validation MAE', color='green')
plt.title("LSTM Training History")
plt.xlabel("Epochs")
plt.ylabel("Loss / MAE")
plt.legend()
plt.show()

sample_index = 5
predicted = lstm_model.predict(x_test[sample_index:sample_index+1])[0, :, 0]
actual = y_test[sample_index, :, 0]
input_window = x_test[sample_index, :, 0]

full_predicted = np.concatenate([input_window, predicted])
full_actual = np.concatenate([input_window, actual])

plt.figure(figsize=(14, 6))
plt.plot(range(len(full_actual)), full_actual, label="Actual", linewidth=2)
plt.plot(range(len(full_predicted)), full_predicted, label="Predicted", linestyle='--')
plt.axvline(x=input_size - 1, color='gray', linestyle=':', label='Prediction Start')
plt.title("LSTM Prediction (Input Window + Predicted Future)")
plt.xlabel("Timestep")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
import matplotlib.dates as mdates
```

```python
def mode_agg(series):
    return series.mode()[0] if not series.mode().empty else np.nan

def create_sequences(data, target, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(target[i + seq_length])
    return np.array(X), np.array(y)
```

```python
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')

numeric_columns = df.select_dtypes(include=['number']).columns.tolist()
string_columns = df.select_dtypes(include=['object']).columns.tolist()

numeric_resampled = df[numeric_columns].resample('h').mean().reset_index()
string_resampled = df[string_columns].resample('h').agg(mode_agg).reset_index()
hourly_df = pd.merge(numeric_resampled, string_resampled, on='time')

hourly_df['hour'] = hourly_df['time'].dt.hour
hourly_df['day_of_week'] = hourly_df['time'].dt.dayofweek
target = 'House overall [kW]'
hourly_df = hourly_df.dropna(subset=[target])

categorical_columns = ['icon', 'summary']
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_data = encoder.fit_transform(hourly_df[categorical_columns])
encoded_columns = encoder.get_feature_names_out(categorical_columns)
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)
hourly_df = hourly_df.drop(columns=categorical_columns)
hourly_df = pd.concat([hourly_df, encoded_df], axis=1)
```

```python
appliance_columns = [
    'Dishwasher [kW]', 'Furnace 1 [kW]', 'Furnace 2 [kW]', 'Home office [kW]', 'Fridge [kW]',
    'Wine cellar [kW]', 'Garage door [kW]', 'Kitchen 12 [kW]', 'Kitchen 14 [kW]',
    'Kitchen 38 [kW]', 'Barn [kW]', 'Well [kW]', 'Microwave [kW]', 'Living room [kW]'
]
hourly_df['total_appliance_power'] = hourly_df[appliance_columns].sum(axis=1)

additional_features = ['temperature', 'humidity', 'visibility', 'windSpeed', 'hour', 'day_of_week']
selected_features = [col for col in appliance_columns + ['total_appliance_power'] + additional_features + list(encoded_columns) if col in hourly_df.columns]
```

```python
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
scaled_data = scaler_X.fit_transform(hourly_df[selected_features])
y = hourly_df[[target]].values
scaled_y = scaler_y.fit_transform(y)
```

```python
n_cols = 3
n_rows = int(np.ceil(len(appliance_columns) / n_cols))
fig, axs = plt.subplots(n_rows, n_cols, figsize=(18, 3 * n_rows),)
axs = axs.flatten()

for i, col in enumerate(appliance_columns):
    axs[i].plot(hourly_df['time'], hourly_df[col])
    axs[i].set_title(col)
    axs[i].set_ylabel("kW")
    axs[i].xaxis.set_major_locator(mdates.MonthLocator())
    axs[i].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
    axs[i].tick_params(axis='x', rotation=45)

for j in range(i + 1, len(axs)):
    axs[j].axis('off')

fig.tight_layout()
plt.show()
```

```python
SEQ_LENGTH = 48
train_size = int(len(scaled_data) * 0.8)
val_size = int(len(scaled_data) * 0.1)

train_data = scaled_data[:train_size]
val_data = scaled_data[train_size:train_size + val_size]
test_data = scaled_data[train_size + val_size:]

train_target = scaled_y[:train_size]
val_target = scaled_y[train_size:train_size + val_size]
test_target = scaled_y[train_size + val_size:]

X_train, y_train = create_sequences(train_data, train_target, SEQ_LENGTH)
X_val, y_val = create_sequences(val_data, val_target, SEQ_LENGTH)
X_test, y_test = create_sequences(test_data, test_target, SEQ_LENGTH)
```

```python
INPUT_DIM = X_train.shape[2]
inputs = layers.Input(shape=(SEQ_LENGTH, INPUT_DIM))
x = layers.Bidirectional(layers.LSTM(128, return_sequences=True))(inputs)
x = layers.Dropout(0.3)(x)
x = layers.Bidirectional(layers.LSTM(64))(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(64, activation='relu')(x)
x = layers.Dense(32, activation='relu')(x)
outputs = layers.Dense(1)(x)
model = models.Model(inputs, outputs)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])


history = model.fit(
    X_train, y_train,
    epochs=50,
    validation_data=(X_val, y_val),
    batch_size=64,
    verbose=1
)
```

```python
predictions = model.predict(X_test)
val_predictions = model.predict(X_val)
actual_kW = scaler_y.inverse_transform(y_test.reshape(-1, 1))
predicted_kW = scaler_y.inverse_transform(predictions)
val_actual_kW = scaler_y.inverse_transform(y_val.reshape(-1, 1))
val_predicted_kW = scaler_y.inverse_transform(val_predictions)
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100
def mean_bias_error(y_true, y_pred):
    return np.mean(y_pred - y_true)
def evaluate_forecasts(y_true, y_pred, label='Test'):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    mbe = mean_bias_error(y_true, y_pred)
    print(f"\n{label} Metrics:")
    print(f"MAE:  {mae:.4f}")
    print(f"MSE:  {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²:   {r2:.4f}")
    print(f"MAPE: {mape:.2f}%")
    print(f"MBE:  {mbe:.4f}")
    return mae, mse, rmse, r2, mbe
```

```python

evaluate_forecasts(actual_kW, predicted_kW, label='Test')
evaluate_forecasts(val_actual_kW, val_predicted_kW, label='Validation')

plt.figure(figsize=(14, 6))
plt.plot(actual_kW, label='Actual', alpha=0.7)
plt.plot(predicted_kW, label='Predicted', alpha=0.7, linestyle='--')
plt.title('Test Set - Prediction vs Actual')
plt.xlabel('Time Steps')
plt.ylabel('House overall [kW]')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

history_dict = history.history
epochs = range(1, len(history_dict['loss']) + 1)

plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, history_dict['loss'], label='Training Loss')
plt.plot(epochs, history_dict['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Loss over Epochs')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs, history_dict['mae'], label='Training MAE')
plt.plot(epochs, history_dict['val_mae'], label='Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('MAE over Epochs')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def create_sequences(data, target, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(target[i + seq_length])
    return np.array(X), np.array(y)
```

```python

# resampling for an hour...
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)
hourly_df = df['House overall [kW]'].resample('h').mean().reset_index()
```

```python
# need to do normalization to avoid exploding/vanishing gradients

scaler = MinMaxScaler()
scaled = scaler.fit_transform(hourly_df[['House overall [kW]']])
```

```python
# 80/10/10

# input=12 hours of past data
# forecast= 3 hours
# jump 15 horus

total = len(scaled)
train_end = int(0.8 * total)
val_end = int(0.9 * total)

train_data = scaled[:train_end]
val_data = scaled[train_end:val_end]
test_data = scaled[val_end:]

def sequences(data, input_len=20, forecast_len=4, jump=24):
    X, y = [], []
    for i in range(0, len(data) - input_len - forecast_len + 1, jump):
        X.append(data[i:i+input_len])
        y.append(data[i+input_len:i+input_len+forecast_len].flatten())
    return np.array(X), np.array(y)

input_len = 12
forecast_len = 3
jump = 15

X_train, y_train = sequences(train_data, input_len, forecast_len, jump)
X_val, y_val = sequences(val_data, input_len, forecast_len, jump)
X_test, y_test = sequences(test_data, input_len, forecast_len, jump)
```

```python
model = Sequential([
    Input(shape=(X_train.shape[1], X_train.shape[2])),
    LSTM(64),
    Dense(3)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(
    X_train, y_train,
    epochs=100,
    validation_data=(X_val, y_val)
)

predictions = model.predict(X_test)
predicted_kW = scaler.inverse_transform(predictions)
actual_kW = scaler.inverse_transform(y_test)

val_predictions = model.predict(X_val)
val_actual = scaler.inverse_transform(y_val)
val_predicted = scaler.inverse_transform(val_predictions)
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100

def mean_bias_error(y_true, y_pred):
    return np.mean(y_pred - y_true)

def evaluate_forecasts(y_true, y_pred, label='Test'):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    mbe = mean_bias_error(y_true, y_pred)
    print(f"\n{label} Metrics:")
    print(f"MAE:  {mae:.4f}")
    print(f"MSE:  {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²:   {r2:.4f}")
    print(f"MAPE: {mape:.2f}%")
    print(f" MBE:  {mbe:.4f}")
    return mae, mse, rmse, r2, mbe

# test and val sets..
mae, mse, rmse, r2, mbe = evaluate_forecasts(actual_kW, predicted_kW)
val_mae, val_mse, val_rmse, val_r2, val_mbe = evaluate_forecasts(val_actual, val_predicted, label='Validation')



fig, axs = plt.subplots(1, 2, figsize=(14, 5))
axs[0].plot(history.history['loss'], label='Train Loss (MSE)')
axs[0].plot(history.history['val_loss'], label='Val Loss (MSE)', linestyle='--')
axs[0].set_title('Loss (MSE)')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(history.history['mae'], label='Train MAE')
axs[1].plot(history.history['val_mae'], label='Val MAE', linestyle='--')
axs[1].set_title('MAE')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Mean Absolute Error')
axs[1].grid(True)
axs[1].legend()

plt.suptitle('Training and Validation Metrics')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# to see numbers more easily...
metric_names = ["MAE", "MSE", "RMSE", "R²", "MBE"]
test_metrics = [mae, mse, rmse, r2, mbe]
val_metrics  = [val_mae, val_mse, val_rmse, val_r2, val_mbe]
x = np.arange(len(metric_names))
width = 0.35
plt.figure(figsize=(12, 5))
plt.bar(x - width/2, val_metrics, width, label='Validation')
plt.bar(x + width/2, test_metrics, width, label='Test')
plt.xticks(x, metric_names, rotation=45)
plt.ylabel('Metric Value')
plt.title('Validation vs Test Metrics')
plt.grid(True, axis='y')
plt.legend()
plt.tight_layout()
plt.show()

'''
brief conclusions on simple model...
Mean Absolute Error (abs diff) - higher on test
Mean Squared Error (avg of squared) -
Root Mean Squared Error (root of above)...
Coefficient of Determination for explaining  variation in the data... -
Mean Absolute Percentage Error - kinda high on test metrics...
Mean Bias Error
'''

```
