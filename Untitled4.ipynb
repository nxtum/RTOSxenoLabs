```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras import layers, models
import tensorflow as tf
import itertools
from tqdm import tqdm
import os

def evaluateForecasts(yTrue, yPred):
    mae = mean_absolute_error(yTrue, yPred)
    mse = mean_squared_error(yTrue, yPred)
    rmse = np.sqrt(mse)
    r2 = r2_score(yTrue, yPred)
    mape = np.mean(np.abs((yTrue - yPred) / (yTrue + 1e-8))) * 100
    mbe = np.mean(yPred - yTrue)
    return mae, mse, rmse, r2, mape, mbe

def createSequences(data, target, seqLen):
    X, y = [], []
    for i in range(len(data) - seqLen):
        X.append(data[i:i + seqLen])
        y.append(target[i + seqLen])
    return np.array(X), np.array(y)

def buildModel(units=64, numLayers=1, bidirectional=False, dropout=0.2,
               activation='tanh', denseLayers=[64, 32], optimizer='adam'):
    inp = layers.Input(shape=(SEQ_LENGTH, XTrain.shape[2]))
    x = inp
    for i in range(numLayers):
        returnSeq = (i < numLayers - 1)
        lstm = layers.LSTM(units, activation=activation, return_sequences=returnSeq)
        x = layers.Bidirectional(lstm)(x) if bidirectional else lstm(x)
        x = layers.Dropout(dropout)(x)
    for d in denseLayers:
        x = layers.Dense(d, activation='relu')(x)
    out = layers.Dense(1)(x)
    model = models.Model(inputs=inp, outputs=out)

    optimizers = {
        'adam': tf.keras.optimizers.Adam(),
        'sgd': tf.keras.optimizers.SGD(),
        'rmsprop': tf.keras.optimizers.RMSprop(),
        'nadam': tf.keras.optimizers.Nadam()
    }

    model.compile(optimizer=optimizers[optimizer], loss='mse', metrics=['mae'])
    return model


def runTrial(params):
    config = dict(zip(gridKeys, params))
    model = buildModel(**{k: config[k] for k in config if k not in ['batchSize', 'epochs']})
    history = model.fit(XTrain, yTrain,
                        validation_data=(XVal, yVal),
                        epochs=config['epochs'],
                        batch_size=config['batchSize'],
                        verbose=0)
    valPred = model.predict(XVal)
    valPredInv = yScaler.inverse_transform(valPred)
    yValInv = yScaler.inverse_transform(yVal)
    mae, mse, rmse, r2, mape, mbe = evaluateForecasts(yValInv, valPredInv)

    return {
        **config,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2,
        'MAPE': mape,
        'MBE': mbe,
        'model': model,
        'history': history.history,
        'valPred': valPredInv,
        'yVal': yValInv
    }


df = pd.read_csv("HomeC.csv", low_memory=False)[:-1]
df['time'] = pd.to_numeric(df['time'], errors='coerce')
df.dropna(subset=['time'], inplace=True)
df['time'] = pd.to_datetime(df['time'], unit='s')
df['time'] = pd.date_range('2016-01-01 05:00', periods=len(df), freq='min')
df.set_index('time', inplace=True)

def modeAgg(series): return series.mode()[0] if not series.mode().empty else np.nan

numCols = df.select_dtypes(include='number').columns.tolist()
strCols = df.select_dtypes(include='object').columns.tolist()
numResampled = df[numCols].resample('h').mean().reset_index()
strResampled = df[strCols].resample('h').agg(modeAgg).reset_index()
dfHourly = pd.merge(numResampled, strResampled, on='time')
dfHourly['hour'] = dfHourly['time'].dt.hour
dfHourly['dayOfWeek'] = dfHourly['time'].dt.dayofweek

targetCol = 'House overall [kW]'
dfHourly.dropna(subset=[targetCol], inplace=True)

encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded = encoder.fit_transform(dfHourly[['icon', 'summary']])
encodedCols = encoder.get_feature_names_out(['icon', 'summary'])
dfEncoded = pd.DataFrame(encoded, columns=encodedCols)
dfHourly = pd.concat([dfHourly.drop(columns=['icon', 'summary']), dfEncoded], axis=1)

appliances = [
    'Dishwasher [kW]', 'Furnace 1 [kW]', 'Furnace 2 [kW]', 'Home office [kW]', 'Fridge [kW]',
    'Wine cellar [kW]', 'Garage door [kW]', 'Kitchen 12 [kW]', 'Kitchen 14 [kW]',
    'Kitchen 38 [kW]', 'Barn [kW]', 'Well [kW]', 'Microwave [kW]', 'Living room [kW]'
]
dfHourly['totalPower'] = dfHourly[appliances].sum(axis=1)
extraFeatures = ['temperature', 'humidity', 'visibility', 'windSpeed', 'hour', 'dayOfWeek']
features = [c for c in appliances + ['totalPower'] + extraFeatures + list(encodedCols) if c in dfHourly.columns]

xScaler = MinMaxScaler()
yScaler = MinMaxScaler()
xScaled = xScaler.fit_transform(dfHourly[features])
yScaled = yScaler.fit_transform(dfHourly[[targetCol]])

SEQ_LENGTH = 48
trainSize = int(len(xScaled) * 0.8)
valSize = int(len(xScaled) * 0.1)

trainX, valX, testX = np.split(xScaled, [trainSize, trainSize + valSize])
trainY, valY, testY = np.split(yScaled, [trainSize, trainSize + valSize])

XTrain, yTrain = createSequences(trainX, trainY, SEQ_LENGTH)
XVal, yVal = createSequences(valX, valY, SEQ_LENGTH)
XTest, yTest = createSequences(testX, testY, SEQ_LENGTH)

paramGrid = {
    'units': [64],
    'numLayers': [1, 2],
    'bidirectional': [True],
    'dropout': [0.2],
    'activation': ['tanh', 'relu'],
    'denseLayers': [[64, 32]],
    'optimizer': ['adam'],
    'batchSize': [32, 64],
    'epochs': [50, 100, 150]
}

gridKeys = list(paramGrid.keys())
gridCombos = list(itertools.product(*paramGrid.values()))
results = [runTrial(params) for params in tqdm(gridCombos, desc="Running Grid Search")]

resultsClean = pd.DataFrame([{k: v for k, v in r.items() if k not in ['model', 'history', 'valPred', 'yVal']} for r in results])
resultsClean.to_csv("grid_search_results.csv", index=False)
resultsClean.to_excel("grid_search_results.xlsx", index=False)

bestIdx = resultsClean['MAE'].idxmin()
bestResult = results[bestIdx]
bestModel = bestResult['model']
bestModel.save("best_lstm_energy_model.h5")

plt.figure(figsize=(14, 5))
plt.plot(bestResult['yVal'], label='Actual')
plt.plot(bestResult['valPred'], label='Predicted', linestyle='--')
plt.title('Validation Set - Actual vs Predicted')
plt.xlabel('Time Steps')
plt.ylabel('kW')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("val_prediction_plot.png")
plt.show()

hist = bestResult['history']
epochs = range(1, len(hist['loss']) + 1)
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, hist['loss'], label='Train Loss')
plt.plot(epochs, hist['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, hist['mae'], label='Train MAE')
plt.plot(epochs, hist['val_mae'], label='Val MAE')
plt.title('MAE over Epochs')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("training_curves.png")
plt.show()

testPred = bestModel.predict(XTest)
testPredInv = yScaler.inverse_transform(testPred)
yTestInv = yScaler.inverse_transform(yTest)
mae, mse, rmse, r2, mape, mbe = evaluateForecasts(yTestInv, testPredInv)

timestampsTest = dfHourly['time'].iloc[trainSize + valSize + SEQ_LENGTH:trainSize + valSize + SEQ_LENGTH + len(yTestInv)].reset_index(drop=True)
timestampsVal = dfHourly['time'].iloc[trainSize + SEQ_LENGTH:trainSize + SEQ_LENGTH + len(bestResult['yVal'])].reset_index(drop=True)

dfVal = pd.DataFrame({
    "Set": "Validation",
    "Timestamp": timestampsVal,
    "Actual_kW": bestResult['yVal'].flatten(),
    "Predicted_kW": bestResult['valPred'].flatten()
})

dfTest = pd.DataFrame({
    "Set": "Test",
    "Timestamp": timestampsTest,
    "Actual_kW": yTestInv.flatten(),
    "Predicted_kW": testPredInv.flatten()
})

dfMetrics = pd.DataFrame([
    {"Set": "Validation", **{k: bestResult[k] for k in ['MAE', 'MSE', 'RMSE', 'R2', 'MAPE', 'MBE']}},
    {"Set": "Test", "MAE": mae, "MSE": mse, "RMSE": rmse, "R2": r2, "MAPE": mape, "MBE": mbe}
])

with pd.ExcelWriter("final_predictions_and_metrics.xlsx", engine='openpyxl') as writer:
    dfVal.to_excel(writer, sheet_name='Validation_Predictions', index=False)
    dfTest.to_excel(writer, sheet_name='Test_Predictions', index=False)
    dfMetrics.to_excel(writer, sheet_name='Evaluation_Metrics', index=False)

```
